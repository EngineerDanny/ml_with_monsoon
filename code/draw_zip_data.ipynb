```python
import os
from urllib.request import urlretrieve
import pandas as pd
import numpy as np
import plotnine as p9


# use pandas.read_csv to read the compressed CSV data file into python as a DataFrame
digit_df = pd.read_csv("/projects/genomic-ml/da2343/ml_project_3/data/zip.test.gz",
                       compression="gzip",
                       sep=" ", header=None)
```

```python
digit_df
```

```python
raw_digit_array = digit_df.to_numpy()
# remove the first column of digit_array
digit_array = raw_digit_array[:, 1:]
init_index = 0
n_pixels = 16
index_vec = np.arange(n_pixels)
image_one_df = pd.DataFrame({
    "col": np.tile(index_vec, n_pixels),
    "row": -np.repeat(index_vec, n_pixels),
    "intensity": digit_array[init_index, :]
})
low_color = "#000000"
high_color = "#FFFFFF"
image_one_label = digit_df.iloc[init_index, 0]
```

```python
gg = p9.ggplot() +\
    p9.ggtitle("Label : " + str(image_one_label)) +\
    p9.geom_raster(
        p9.aes(
            x="col",
            y="row",
            fill="intensity",
        ),
        data=image_one_df) +\
    p9.scale_fill_gradient(low=low_color, high=high_color)
gg.save("/projects/genomic-ml/da2343/ml_project_3/figures/digit_image.png")
```

```python


list_of_data_frames = []
for init_index in range(20):
    image_df = pd.DataFrame({
        "observation": init_index,
        "label": digit_df.iloc[init_index, 0],
        "intensity": digit_array[init_index, :],
        "row": -np.repeat(index_vec, n_pixels),
        "col": np.tile(index_vec, n_pixels)
    })

    list_of_data_frames.append(image_df)

# concatenate all the dataframes in the list into a single dataframe
all_images_df = pd.concat(list_of_data_frames)
print(all_images_df)

# make a facetted plot with one panel per image
gg = p9.ggplot() +\
    p9.geom_raster(
        p9.aes(
            x="col",
            y="row",
            fill="intensity",
        ),
        data=all_images_df) +\
    p9.scale_fill_gradient(low=low_color, high=high_color) +\
    p9.facet_wrap(["observation", "label"], labeller="label_both")
```

```python
gg.save("/projects/genomic-ml/da2343/ml_project_3/figures/digit_images_facetted.png", dpi=1000)
```
